```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
X1 = np.array([30, 30, 40, 40])
Y1 = np.array([37, 47, 50, 60])

X2 = np.array([30, 30, 40, 40, 20, 20, 50, 50])
Y2 = np.array([37, 47, 50, 60, 25, 35, 62, 72])
    
X3 = np.array([30, 30, 40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
Y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]) 
```

```python
plt.scatter(X1, Y1)
plt.show()

np.corrcoef(X1, Y1) # коэффициент корреляции (какая-то линейная зависимость существует)
```

```python
from sklearn.linear_model import LinearRegression

X1 = X1.reshape(-1, 1)      # строим двумерный массив
model = LinearRegression()  # строим модель
model.fit(X1, Y1)

model.intercept_, model.coef_

plt.plot(X1, X1 * model.coef_[0] + model.intercept_, color = 'red')

plt.scatter(X1, Y1)
plt.show()
```

```python
X2 = X2.reshape(-1, 1)      # строим двумерный массив
model = LinearRegression()  # строим модель
model.fit(X2, Y2)

model.intercept_, model.coef_

plt.plot(X2, X2 * model.coef_[0] + model.intercept_, color = 'red')

plt.scatter(X2, Y2)
plt.show()
```

```python
X3 = X3.reshape(-1, 1)      # строим двумерный массив
model = LinearRegression()  # строим модель
model.fit(X3, Y3)

model.intercept_, model.coef_

plt.plot(X3, X3 * model.coef_[0] + model.intercept_, color = 'red')

plt.scatter(X3, Y3)
plt.show()
```
